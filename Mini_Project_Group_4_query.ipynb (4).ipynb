1. Identify the Most Popular Cuisines:

Insight: Discover the most popular cuisines to assess potential competition and demand.

```python
query = """
SELECT categories, COUNT(*) AS restaurant_count
FROM business
WHERE categories LIKE '%Restaurant%'
GROUP BY categories
ORDER BY restaurant_count DESC;
"""
pop_cuisine = pd.read_sql(query, engine)
pop_cuisine.head(10)
```

2. TOP 10 rated cities:

Insight: Discover cities with highest ratings. It is good for us to know competition environment and competitors.

```python
query = """
SELECT city, state, AVG(stars) AS avg_rating, COUNT(*) AS total_restaurants
FROM business
WHERE categories LIKE '%Restaurant%'
GROUP BY city, state
HAVING COUNT(*) > 100
ORDER BY avg_rating DESC, total_restaurants DESC;
"""
top_rated_cities = pd.read_sql(query, engine)
top_rated_cities.head(10)
```

3. Top 10 reviewed cities

Insights: Discover cities with highest reviews. More reviews mean the restaurants in that city have more customers than others.

```python
query = """
SELECT city, state, SUM(review_count) AS total_reviews
FROM business
WHERE categories
LIKE '%Restaurant%'
GROUP BY city, state
ORDER BY total_reviews DESC;
"""
top_review_cities = pd.read_sql(query, engine)
top_review_cities.head(10)
```

4. Average Rating of Restaurants in the City:

Insight: We choose Indianapolis as our destination because it is in the middle of the top 10 cities has most reviews. Understand the general standard for restaurant ratings to establish a benchmark for quality.

```python
query = """
SELECT AVG(stars) AS avg_rating
FROM business
WHERE city = 'Indianapolis' AND categories LIKE '%Restaurant%';
"""

Avg_stars = pd.read_sql(query, engine)

# Display the first few rows of the resulting DataFrame
print(Avg_stars.head())
```

5. Top-Rated Restaurants in a Specific Cuisine:

Insight: Identify popular restaurants in a specific cuisine type to analyze successful competitors.

```python
query = """
SELECT TOP 10 name, stars, review_count
FROM business
WHERE city = 'Indianapolis' AND categories LIKE '%Restaurant%' AND stars >= 4
ORDER BY stars DESC, review_count DESC;
"""
top_rated = pd.read_sql(query, engine)

# Display the first few rows of the resulting DataFrame
print(top_rated.head())
```

7. Review Count vs. Average Rating Analysis:
Insight: Evaluate if high ratings correlate with more reviews, indicating strong customer engagement.

```python
query = """
SELECT stars, COUNT(review_count) AS review_count
FROM business
WHERE city = 'Indianapolis' AND categories LIKE '%Restaurant%'
GROUP BY stars
ORDER BY stars DESC;
"""

review_vs_rate = pd.read_sql(query, engine)
review_vs_rate
```

8. Identify Busy Times with High Demand for Certain Cuisine Types:

Insight: Recognize peak hours for certain cuisines, helping optimize staffing and menu availability.

```python
query = """
SELECT
    b.categories,
    DATEPART(hour, c.date) AS peak_hour,
    COUNT(*) AS popularity
FROM
    business b
JOIN
    checkin c ON b.business_id = c.business_id
WHERE
    b.city = 'Indianapolis' AND b.categories LIKE '%Restaurant%'
GROUP BY
    b.categories, DATEPART(hour, c.date)
ORDER BY
    popularity DESC;
"""

busy_time = pd.read_sql(query, engine)
busy_time.head(10)
```

9. Most active users with their ratings and reviews count

Insights: The list of those users who have the highest review count and their average ratings given. Be aware of their comments on other restuarants may help us improve or avoid certain amennities, cuisines, or services.

```python
query="""
SELECT
    u.user_id AS user_id,
    u.name AS user_name,
    COUNT(r.business_id) AS review_count,
    AVG(r.stars) AS average_rating_given
FROM
    [user] u
JOIN
    review r ON u.user_id = r.user_id
GROUP BY
    u.user_id, u.name
ORDER BY
    review_count DESC
OFFSET 0 ROWS FETCH NEXT 10 ROWS ONLY;
"""

active_user = pd.read_sql(query, engine)
active_user.head(10)
```

10. Identify Competitor Density by Cuisine Type.
**(Decision-making)**
Helps determine which cuisines are oversaturated or underserved in the market, enabling a new restaurant to choose a cuisine type with less competition.



```python
query="""
SELECT categories, COUNT(business_id) AS restaurant_count
FROM business
WHERE is_open = 1 AND categories LIKE '%Restaurant%'
GROUP BY categories
ORDER BY restaurant_count DESC;
"""

competitor_dens = pd.read_sql(query, engine)
competitor_dens.head(10)
```
